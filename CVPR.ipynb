```python
import numpy as np
# First create some toy data:
x = np.arange(-10, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure()
plt.plot(x,y1, x,y2)
plt.legend(['Sine','Cosine'])
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('Sine')
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Cosine')
plt.show()
```
